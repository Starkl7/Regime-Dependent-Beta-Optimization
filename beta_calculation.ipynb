```python
import pandas as pd
import pickle
```

```python
df_equities = pd.read_excel('Datasets/updated3.2_All_Equities_Cleaned.xlsx')
display(df_equities.head())
df_equities.to_csv('Datasets/All_Equities_Cleaned.csv', index=False)
```

```python
df_equity = pd.read_csv('Datasets/All_Equities_Cleaned.csv', parse_dates=['Dates'])
df_market = pd.read_csv('Datasets/MSCI_Data.csv', parse_dates=['Date'])
```

```python
display(df_market.head())
```

```python
print(df_market.info())
```

```python
df_market['Date'] = df_market['Date'].dt.to_period('M')
df_equity['Dates'] = df_equity['Dates'].dt.to_period('M')
```

```python
df_market.set_index('Date', inplace=True)
df_equity.set_index('Dates', inplace=True)
df_merged = df_equity.join(df_market, how='inner')
display(df_merged.head())
```

```python
returns_df = df_merged.pct_change()
display(returns_df.head())
```

```python
print(list(df_market.columns))
```

```python
targets = list(df_market.columns)

equities = [col for col in returns_df.columns if col not in targets]

#Rolling beta for 5 yrs (60 months)
rolling_var = returns_df[targets].rolling(window=60).var()
all_betas = {}

for equity in equities:
    equity_betas = pd.DataFrame(index=returns_df.index)
    
    for target in targets:
        rolling_cov = returns_df[equity].rolling(window=60).cov(returns_df[target])
        equity_betas[f'Beta_{target}'] = rolling_cov / rolling_var[target]
        
    equity_betas.dropna(inplace=True)
    all_betas[equity] = equity_betas
```

```python
def clean_key_names(dict_of_dfs):
    cleaned_dict = {}
    for key, df in dict_of_dfs.items():
        cleaned_key = key.split(' ')[0]
        cleaned_dict[cleaned_key] = df
    return cleaned_dict
```

```python
all_betas_cleaned = clean_key_names(all_betas.copy())
print(all_betas_cleaned.keys())
```

```python
display(all_betas_cleaned['AAPL'].head())
```

```python
with open('Datasets/all_betas_cleaned.pkl', 'wb') as f:
    pickle.dump(all_betas_cleaned, f)
```
